```python
import pandas as pd
#Importing
fp_train=pd.read_csv("Final_startup_funding.csv")#Importing the Full file
fp_test= fp_train[0:int(len(fp_train)*0.20)]#Importing a test file
industry_test=fp_test['Industry Vertical']#DataFrame consisting of only Industry Vertical
industry_train=fp_train['Industry Vertical']#DataFrame consisting of only Industry Vertical
city_test=fp_test['City  Location']#DataFrame consisting of only City  Location
city_train=fp_train['City  Location']#DataFrame consisting of only City  Location
amount_test=fp_test['Amount']#DataFrame consisting of only Amount
amount_train=fp_train['Amount']#DataFrame consisting of only Amount
train=[]
test=[]
for i in range(len(amount_train)):#Storing city,industry vertical,amount from training set features  in different lists 
    lis=[city_train[i],industry_train[i],amount_train[i]]
    train.append(lis)#nested List called train
for i in range(len(amount_test)):
    lis=[city_test[i],industry_test[i],amount_test[i]]#Storing city,industry vertical,amount from testing set features  in different lists 
    test.append(lis)#nested List called test

amount=[]
predicted=[]#Empty list for storing prediction amount
for x_test in test:
    for x_train in train:
        if(x_test[0]==x_train[0] and x_test[1]==x_train[1]):
            amount.append(int(x_train[2]))
    summation= sum(amount)/len(amount)
    predicted.append(summation)
error=0;

for i in range(0,len(amount_test)):
    if abs(predicted[i]-int(amount_test[i]))/int(amount_test[i]) <= 56:#Removing Outliers where the difference is greater than 56%
       error+=abs((predicted[i]-int(amount_test[i]))/int(amount_test[i]))
    
MEAN_ABSOLUTE_PERCENTAGE_ERROR=error/len(amount_test)#Root error
print("error=",MEAN_ABSOLUTE_PERCENTAGE_ERROR,"%")#Calculating error
print("accuracy=",100-MEAN_ABSOLUTE_PERCENTAGE_ERROR,"%")#Calculating accuracy
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
from statistics import mean
from sklearn.linear_model import LinearRegression



fullData = pd.read_csv("/Users/jeshwanth/Desktop/Final_startup_funding.csv");#Importing file
#Dropping irrelevant columns
fullData = fullData.drop('Startup Name', 1);
fullData = fullData.drop('Investors Name', 1);
fullData = fullData.drop('Date dd/mm/yyyy', 1);



train, test = fullData[0:int(len(fullData)*0.7)], fullData[int(len(fullData)*0.7):]#Splitting the dataset to tran and test wrt a ratio of 70:30

train['Type']='Train';#Classifying the values if they are used for training by introducing a new feature called type
test['Type']='Test';#Classifying the values if they are used for testing by introducing a new feature called type

ID_col = ['Sr No']#Columns that only have id's
target_col = ["Amount"]#Our target dataframe

other_col=['Type'] #Test and Train Data set identifier
cols = ['Industry Vertical','City  Location','InvestmentnType'];#All categorial value columns 

for var in cols:#Label encoding the categorical columns
 number = LabelEncoder()
 fullData[var] = number.fit_transform(fullData[var].astype('str'))#Converting to a integer

y= fullData.Amount#Dataframe having only Amount
 

x=fullData.drop('Amount',axis=1)#Dataframe having all features except amount
 
#splitting for testing and traing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
 
#Creating the linear model
model = LinearRegression()
#Fitting the linear model with relevant features
model.fit(x_train,y_train);
#Predicting using x_test which we got by splitting
y_pred = model.predict(x_test)

y_pred = np.array(y_pred)#Converting to a numpy array
y_pred = [abs(k) for k in y_pred]#Converting negative to positive numbers





#Root Mean Squared Error
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)));

#Mean Absolute Error
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred));

#Mean Squared Error
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred));





```
